```python
import pickle
from tqdm import tqdm
```

```python
url  = "https://comic.naver.com/api/article/list?titleId={}&page=1"
```

```python

with open("./webtoon_id.pkl", "rb") as f:
    title_ids = pickle.load(f)
```

```python
title_ids
```

```python
pages_list=[]

for i in tqdm(title_ids):
    # print(type(i))
    r = requests.get(url.format(i))
    try :
        page = r.json()['articleList'][0]['no']
        # print(page)
        pages_list.append(page)
    except TypeError as te:
        print(te)
```

```python
pages_list.__len__()
```
